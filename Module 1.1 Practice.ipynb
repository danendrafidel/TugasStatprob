**1.Diketahui tinggi badan siswa kelas A adalah 100.5, 99.7, 101.0, 101.3, 99.9, 98.9, 165.0, 102.2, 98.7, 102.9, dan 100.0 . Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?**

*A.Mean*

```python
# Inisialisasi dataset
data = [100.5, 99.7, 101.0, 101.3, 99.9, 98.9, 165.0, 102.2, 98.7, 102.9, 100.0]
print(f"Data: {data}")

# Jumlah nilai dalam dataset 
jumlah_nilai = sum(data)
print(f"Jumlah semua nilai: {jumlah_nilai}")

# Jumlah total data dalam dataset 
jumlah_total_data = len(data)
print(f"Jumlah total data: {jumlah_total_data}")

# Jumlah semua nilai dalam dataset 
mean_value = jumlah_nilai / jumlah_total_data
print(f"Rata-rata: {mean_value}")

import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='blue')

# Menambahkan garis vertikal di mean
plt.axvline(x=mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')

# Menambahkan label dan judul
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Mean')

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```

*B.Median*

```python
# Inisialisasi dataset
data = [100.5, 99.7, 101.0, 101.3, 99.9, 98.9, 165.0, 102.2, 98.7, 102.9, 100.0]
print(f"Data: {data}")

# Mengurutkan data dari terkecil hingga terbesar
data_terurut = sorted(data)
print(f"Data terurut: {data_terurut}")

# Jumlah total data dalam dataset 
n = len(data_terurut)
print(f"Jumlah total data: {n}")

index_tengah = n // 2  # double slash // digunakan untuk pembagian pembulatan ke bawah

# Jika jumlah data genap, median adalah rata-rata dari dua nilai tengah 
if n % 2 == 0:
    print(f"Jumlah total data Genap")
    median1 = data_terurut[index_tengah]
    median2 = data_terurut[index_tengah - 1]
    print(f"Nilai tengah: {median1} {median2}")
    nilai_median = (median1 + median2) / 2
    
# Jika jumlah data ganjil, median adalah nilai tengah
else:
    print(f"Jumlah total data Ganjil")
    nilai_median = data_terurut[index_tengah]
    print(f"Nilai tengah: {nilai_median}")

print(f"Median: {nilai_median}")

import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='yellow')

# Menambahkan garis vertikal di median
plt.axvline(x=nilai_median, color='red', linestyle='dashed', linewidth=1, label=f'Median = {nilai_median:.2f}')

# Menambahkan label dan judul
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Median')

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.show()

```

*C.Modus*

```python
# Impor library Counter untuk menghitung modus
from collections import Counter

# Inisialisasi dataset
data = [100.5, 99.7, 101.0, 101.3, 99.9, 98.9, 165.0, 102.2, 98.7, 102.9, 100.0]
print(f"Data: {data}")

counter = Counter(data)
data_modus = dict(counter)  # ubah ke tipe data dictionary
maks_counter = max(list(counter.values()))  # mencari data yang memiliki counter paling banyak
print(f"Counter Maksimum: {maks_counter}")

nilai_modus = []
for k, v in data_modus.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
        nilai_modus.append(k)

print(f"Modus: {nilai_modus}")

import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Menambahkan garis vertikal di modus
plt.axvline(x=nilai_modus[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {nilai_modus[0]:.2f}')
plt.axvline(x=nilai_modus[1], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 2 = {nilai_modus[1]:.2f}')

# Menambahkan label dan judul
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Modus')

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```

```python
# Inisialisasi dataset
data = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]
print(f"Data: {data}")

# Jumlah nilai dalam dataset 
jumlah_nilai = sum(data)
print(f"Jumlah nilai: {jumlah_nilai}")

# Banyak data dalam dataset 
banyak_data = len(data)
print(f"Banyak data: {banyak_data}")

# Jumlah semua nilai dalam dataset 
nilai_rata_rata = jumlah_nilai / banyak_data
print(f"Rata-rata: {nilai_rata_rata}")

import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='red')

# Menambahkan garis vertikal di rata-rata
plt.axvline(x=nilai_rata_rata, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {nilai_rata_rata:.2f}')

# Menambahkan label dan judul
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Mean')

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```
