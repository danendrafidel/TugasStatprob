**<span style="font-size:1.5em;">Teorema Bayes</span>**

```python
# Fungsi untuk menghitung Teorema Bayes
def teorema_bayes(prior_prob_A, prob_B_diketahui_A, prob_B):
    prob_A_diketahui_B = (prob_B_diketahui_A * prior_prob_A) / prob_B
    return prob_A_diketahui_B

# Contoh data
prior_prob_A = 0.3  # Probabilitas awal A
prob_B_diketahui_A = 0.8  # Probabilitas B jika A terjadi
prob_B = 0.5  # Probabilitas awal B

# Menghitung Teorema Bayes
hasil_bayes = teorema_bayes(prior_prob_A, prob_B_diketahui_A, prob_B)

# Menampilkan hasil
print("Probabilitas A jika B terjadi:", hasil_bayes)
```
